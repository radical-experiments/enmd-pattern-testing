```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import math
import numpy as np
import glob
import os
%matplotlib
```

```python
trials = 3
cores = [24,48,96,192]
```

```python
def extract_time_date(t1,t2,t3=None,t4=None):
    
    try:
        t1 = datetime.datetime.strptime(t1, "%Y-%m-%d %H:%M:%S.%f")
    except:
        t1 = datetime.datetime.strptime(t1, "%Y-%m-%d %H:%M:%S")
        
    try:
        t2  = datetime.datetime.strptime(t2, "%Y-%m-%d %H:%M:%S.%f")
    except:
        t2  = datetime.datetime.strptime(t2, "%Y-%m-%d %H:%M:%S")
        
    if t3 is not None:
        
        try:
            t3   = datetime.datetime.strptime(t3, "%Y-%m-%d %H:%M:%S.%f")
        except:
            t3   = datetime.datetime.strptime(t3, "%Y-%m-%d %H:%M:%S")
        
        try:
            t4  = datetime.datetime.strptime(t4, "%Y-%m-%d %H:%M:%S.%f")
        except:
            t4  = datetime.datetime.strptime(t4, "%Y-%m-%d %H:%M:%S")
            
            
        return ((t2-t1)+(t4-t3)).total_seconds()
    
    else:
        return (t2-t1).total_seconds()
```

```python
def extract_time(t1,t2,t3=None,t4=None):
    if t3 is not None:
        t1=float(t1)
        t2=float(t2)
        t3=float(t3)
        t4=float(t4)
        return ((t2-t1)+(t4-t3))
    else:
        t1=float(t1)
        t2=float(t2)
        return (t2-t1)
```

```python
def extract_pat_overhead(df):
    step1_df = df[:4]
    step2_df = df[4:]
    
    start,wait,res,stop = step1_df['timestamp']    
    step1_ov = extract_time_date(start,wait,res,stop)
    
    start,wait,res,stop = step2_df['timestamp']    
    step2_ov = extract_time_date(start,wait,res,stop)
    
    #print step1_ov + step2_ov
    return step1_ov + step2_ov
```

```python
def extract_core_overhead(df):
    alloc_start,alloc_stop,dealloc_start,dealloc_stop = df['timestamp']
    core_ov = extract_time_date(alloc_start,alloc_stop,dealloc_start,dealloc_stop)
    
    return core_ov
```

```python
def extract_timing_info(df1,df2,df3,cores,inst):
    
    step1_data,step2_data = extract_data_time(df1,cores,inst)
    step1_exec,step2_exec = extract_exec_time_pipe(df1,cores,inst)
    sim_sal_exec,ana_exec = extract_exec_time_sal(df2,cores,inst)
    sim_repex_exec,ex_exec = extract_exec_time_repex(df3,cores,inst)
    
    return [step1_data, step2_data],[step1_exec, step2_exec],[sim_sal_exec,ana_exec],[sim_repex_exec,ex_exec]
```

```python
def extract_exec_time_repex(df,cores,inst):
    req_df = df[['Step','Executing','AgentStagingOutputPending']]
    
    sim_df = req_df[:inst]
    ex_df = req_df[inst:]
    
    
    #Step1 extraction
    sim_exec_time  = list()
    for row in sim_df.iterrows():
        step,t1,t2 = row[1:][0]
        if type(t1) is float:
            sim_exec_time.append(extract_time(t1,t2))
        
    sim_exec_time = np.average(sim_exec_time)
    
    #Step2 extraction
    ex_exec_time  = list()
    for row in ex_df.iterrows():
        step,t1, t2 = row[1:][0]
        if type(t1) is float:
            ex_exec_time.append(extract_time(t1,t2))
    
    ex_exec_time = np.average(ex_exec_time)
    
    
    #print step1_data_time,step2_data_time
    return sim_exec_time,ex_exec_time
```

```python
def extract_exec_time_pipe(df,cores,inst):
    req_df = df[['step','Executing','AgentStagingOutputPending']]
    
    step1_df = req_df[:inst]
    step2_df = req_df[inst:]
    
    
    #Step1 extraction
    step1_exec_time  = list()
    for row in step1_df.iterrows():
        step,t1,t2 = row[1:][0]
        step1_exec_time.append(extract_time(t1,t2))
        
    step1_exec_time = np.average(step1_exec_time)
    
    #Step2 extraction
    step2_exec_time  = list()
    for row in step2_df.iterrows():
        step,t1, t2 = row[1:][0]
        step2_exec_time.append(extract_time(t1,t2))
    
    step2_exec_time = np.average(step2_exec_time)
    
    
    #print step1_data_time,step2_data_time
    return step1_exec_time,step2_exec_time
```

```python
def extract_exec_time_sal(df,cores,inst):
    req_df = df[['step','Executing','AgentStagingOutputPending']]
    
    sim_df = req_df[:inst]
    ana_df = req_df[inst:]
    
    
    #Step1 extraction
    sim_exec_time  = list()
    for row in sim_df.iterrows():
        step,t1,t2 = row[1:][0]
        sim_exec_time.append(extract_time(t1,t2))
        
    sim_exec_time = np.average(sim_exec_time)
    
    #Step2 extraction
    ana_exec_time  = list()
    for row in ana_df.iterrows():
        step,t1, t2 = row[1:][0]
        ana_exec_time.append(extract_time(t1,t2))
    
    ana_exec_time = np.average(ana_exec_time)
    
    
    #print step1_data_time,step2_data_time
    return sim_exec_time,ana_exec_time
```

```python
def extract_rp_overhead(df,data,exec_list):
    
    step1_df = df[:4]
    step2_df = df[4:]
    
    start,wait,res,stop = step1_df['timestamp']    
    step1_wait = extract_time_date(wait,res)
    step1_rp_ov = step1_wait - data[0] - exec_list[0]
    
    start,wait,res,stop = step2_df['timestamp']    
    step2_wait = extract_time_date(wait,res)
    step2_rp_ov = step2_wait - data[1] - exec_list[1]
    
    rp_ov = step1_rp_ov + step2_rp_ov
    
    
    #print preloop_rp_ov, sim_rp_ov, ana_rp_ov
    
    return rp_ov
```

```python
def extract_data_time(df,cores,inst):
    req_df = df[['step','StagingInput','Allocating','AgentStagingOutputPending','Done']]
    
    step1_df = req_df[:inst]
    step2_df = req_df[inst:]
    
    
    #Step1 extraction
    step1_data_time  = list()
    for row in step1_df.iterrows():
        step,t1,t2,t3,t4 = row[1:][0]
        step1_data_time.append(extract_time(t1,t2,t3,t4))
        
    step1_data_time = np.average(step1_data_time)
    
    #Step2 extraction
    step2_data_time  = list()
    for row in step2_df.iterrows():
        step,t1,t2,t3,t4 = row[1:][0]
        #print t1,t2,t3,t4
        step2_data_time.append(extract_time(t1,t2,t3,t4))
    
    step2_data_time = np.average(step2_data_time)
    
    
    #print step1_data_time,step2_data_time
    return step1_data_time,step2_data_time
```

```python
def split_file(t,c):
    fname = glob.glob('repex/experiment_iter{0}_p{1}_i{1}/execution*.csv'.format(t,c))
    #print fname
    fname=fname[0]
    try:
        os.remove('enmd_pat_overhead.csv')
    except:
        pass
    try:
        os.remove('execution_profile.csv')
    except:
        pass
    l=0
    with open(fname,'r') as f1:
        for line in f1:
            if (l<7):
                with open('enmd_pat_overhead.csv','a') as f2:
                    f2.write(line)
                l+=1
            else:
                with open('execution_profile.csv','a') as f3:
                    f3.write(line)
                    
    df_pat = pd.read_csv('enmd_pat_overhead.csv',header=0,sep=';')
    df_cu = pd.read_csv('execution_profile.csv',header=0,sep=';',skipinitialspace=True).sort('CU_ID')
        
    return df_pat, df_cu
```

```python
def separate(comp_df,err_df):
    
    # Overhead dataframe
    ov_df = pd.DataFrame(columns=['EnMD Core overhead',
                                'EnMD Pat overhead',
                                'RP overhead'])
    ov_df_err = pd.DataFrame(columns=['EnMD Core overhead',
                                'EnMD Pat overhead',
                                'RP overhead'])
    ov_df['EnMD Core overhead'] = comp_df['EnMD Core overhead']
    ov_df['EnMD Pat overhead'] = comp_df['EnMD Pat overhead']
    ov_df['RP overhead'] = comp_df['RP overhead']
    
    ov_df_err['EnMD Core overhead'] = err_df['EnMD Core overhead']
    ov_df_err['EnMD Pat overhead'] = err_df['EnMD Pat overhead']
    ov_df_err['RP overhead'] = err_df['RP overhead']
    
    
    # Pipeline dataframe
    app_df_pipe = pd.DataFrame(columns=['step_1 execution time',
                                    'step_2 execution time'])
    app_df_pipe_err = pd.DataFrame(columns=['step_1 execution time',
                                    'step_2 execution time'])

    app_df_pipe['step_1 execution time'] = comp_df['step_1 execution time']
    app_df_pipe['step_2 execution time'] = comp_df['step_2 execution time']
    
    app_df_pipe_err['step_1 execution time'] = err_df['step_1 execution time']
    app_df_pipe_err['step_2 execution time'] = err_df['step_2 execution time']
    
    
    # SAL dataframe
    app_df_sal = pd.DataFrame(columns=['simulation execution time',
                                    'analysis execution time'])
    app_df_sal_err = pd.DataFrame(columns=['simulation execution time',
                                    'analysis execution time'])

    app_df_sal['simulation execution time'] = comp_df['sim_sal execution time']
    app_df_sal['analysis execution time'] = comp_df['ana execution time']
    
    app_df_sal_err['simulation execution time'] = err_df['sim_sal execution time']
    app_df_sal_err['analysis execution time'] = err_df['ana execution time']
    
    
    # Repex dataframe
    app_df_repex = pd.DataFrame(columns=['simulation execution time',
                                    'exchange execution time'])
    app_df_repex_err = pd.DataFrame(columns=['simulation execution time',
                                    'exchange execution time'])

    app_df_repex['simulation execution time'] = comp_df['sim_repex execution time']
    app_df_repex['exchange execution time'] = comp_df['ana execution time']
    
    app_df_repex_err['simulation execution time'] = err_df['sim_repex execution time']
    app_df_repex_err['exchange execution time'] = err_df['ana execution time']
    
    
    
    data_df = pd.DataFrame(columns=['data movement'])
    data_df_err = pd.DataFrame(columns=['data movement'])
    data_df['data movement'] = comp_df['data movement']
    data_df_err['data movement'] = err_df['data movement']
    
    split_plot([ov_df,ov_df_err],[app_df_pipe,app_df_pipe_err],[app_df_sal,app_df_sal_err],[app_df_repex,app_df_repex_err],[data_df,data_df_err])
```

```python
#MAIN

comp_df = pd.DataFrame(columns=['EnMD Core overhead',
                                'EnMD Pat overhead',
                                'RP overhead',
                                'step_1 execution time',
                                'step_2 execution time',
                                'sim_sal execution time',
                                'ana execution time',
                                'sim_repex execution time',
                                'exch execution time',
                                'data movement'])

err_df = pd.DataFrame(columns=['EnMD Core overhead',
                                'EnMD Pat overhead',
                                'RP overhead',
                                'step_1 execution time',
                                'step_2 execution time',
                                'sim_sal execution time',
                                'ana execution time',
                                'sim_repex execution time',
                                'exch execution time',
                                'data movement'])


for c in cores:
    
    core_ov_list = list()
    pat_ov_list = list()
    rp_ov_list = list()
    data_list = list()
    step1_exec_list = list()
    step2_exec_list = list()
    sim_sal_exec_list = list()
    ana_exec_list = list()
    sim_repex_exec_list = list()
    exch_exec_list = list()
    
    for t in range(1,trials+1):
        
        #f_core = open('experiment_iter{0}_p{1}_i{1}/enmd_core_overhead.csv'.format(t,c))
        df_core = pd.read_csv('pipe/experiment_iter{0}_p{1}_i{1}/enmd_core_overhead.csv'.format(t,c),header=0,sep=',')
        df_pat = pd.read_csv('pipe/experiment_iter{0}_p{1}_i{1}/enmd_pat_overhead.csv'.format(t,c),header=0,sep=',')
        
        fname = glob.glob('pipe/experiment_iter{0}_p{1}_i{1}/execution*.csv'.format(t,c))[0]
        df1 = pd.read_csv('{0}'.format(fname),header=0,sep=',',skipinitialspace=True)
        
        dfp, df3 = split_file(t,c)
        
        fname = glob.glob('sal/experiment_iter{0}_p{1}_i{1}/execution*.csv'.format(t,c))[0]
        df2 = pd.read_csv('{0}'.format(fname),header=0,sep=',',skipinitialspace=True)
        
        core_ov = extract_core_overhead(df_core)
        pat_ov = extract_pat_overhead(df_pat)
        data, exec1, exec2, exec3 = extract_timing_info(df1,df2,df3,cores=c,inst=c)
        rp_ov = extract_rp_overhead(df_pat,data,exec1)
        
        
        core_ov_list.append(core_ov)
        pat_ov_list.append(pat_ov)
        rp_ov_list.append(rp_ov)
        step1_exec_list.append(exec1[0])
        step2_exec_list.append(exec1[1])
        sim_sal_exec_list.append(exec2[0])
        ana_exec_list.append(exec2[1])
        sim_repex_exec_list.append(exec3[0])
        exch_exec_list.append(exec3[1])
        data_list.append(sum(data))
    
    comp_df.loc['{0}/{1}'.format(c,c)] = [np.average(core_ov_list),
                                          np.average(pat_ov_list),
                                          np.average(rp_ov_list),
                                          np.average(step1_exec_list),
                                          np.average(step2_exec_list),
                                          np.average(sim_sal_exec_list),
                                          np.average(ana_exec_list),
                                          np.average(sim_repex_exec_list),
                                          np.average(exch_exec_list),
                                          np.average(data_list)]
    
    err_df.loc['{0}/{1}'.format(c,c)] = [
                                          np.std(core_ov_list)/math.sqrt(trials),
                                          np.std(pat_ov_list)/math.sqrt(trials),
                                          np.std(rp_ov_list)/math.sqrt(trials),
                                          np.std(step1_exec_list)/math.sqrt(trials),
                                          np.std(step2_exec_list)/math.sqrt(trials),
                                          np.std(sim_sal_exec_list)/math.sqrt(trials),
                                          np.std(ana_exec_list)/math.sqrt(trials),
                                          np.std(sim_repex_exec_list)/math.sqrt(trials),
                                          np.std(exch_exec_list)/math.sqrt(trials),
                                          np.std(data_list)/math.sqrt(trials)
                                         ]
    
    
separate(comp_df,err_df)
```
