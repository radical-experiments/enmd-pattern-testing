```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import math
import numpy as np
import glob
%matplotlib
```

```python
trials = 3
cores = [24,48,96,192]
```

```python
def extract_time_date(t1,t2,t3=None,t4=None):
    
    try:
        t1 = datetime.datetime.strptime(t1, "%Y-%m-%d %H:%M:%S.%f")
    except:
        t1 = datetime.datetime.strptime(t1, "%Y-%m-%d %H:%M:%S")
        
    try:
        t2  = datetime.datetime.strptime(t2, "%Y-%m-%d %H:%M:%S.%f")
    except:
        t2  = datetime.datetime.strptime(t2, "%Y-%m-%d %H:%M:%S")
        
    if t3 is not None:
        
        try:
            t3   = datetime.datetime.strptime(t3, "%Y-%m-%d %H:%M:%S.%f")
        except:
            t3   = datetime.datetime.strptime(t3, "%Y-%m-%d %H:%M:%S")
        
        try:
            t4  = datetime.datetime.strptime(t4, "%Y-%m-%d %H:%M:%S.%f")
        except:
            t4  = datetime.datetime.strptime(t4, "%Y-%m-%d %H:%M:%S")
            
            
        return ((t2-t1)+(t4-t3)).total_seconds()
    
    else:
        return (t2-t1).total_seconds()
```

```python
def extract_time(t1,t2,t3=None,t4=None):
    if t3 is not None:
        t1=float(t1)
        t2=float(t2)
        t3=float(t3)
        t4=float(t4)
        return ((t2-t1)+(t4-t3))
    else:
        t1=float(t1)
        t2=float(t2)
        return (t2-t1)
```

```python
def extract_core_overhead(df):
    alloc_start,alloc_stop,dealloc_start,dealloc_stop = df['timestamp']
    core_ov = extract_time_date(alloc_start,alloc_stop,dealloc_start,dealloc_stop)
    
    return core_ov
```

```python
def extract_pat_overhead(df):
    preloop_df = df[:4]
    sim_df = df[4:10]
    ana_df = df[10:]
    
    start,wait,res,stop = preloop_df['timestamp']
    preloop_ov = extract_time_date(start,wait,res,stop)
    
    start,wait,res,stop,post_start,post_stop = sim_df['timestamp']    
    sim_ov = extract_time_date(start,wait,res,stop) + extract_time_date(post_start,post_stop)
    
    start1,wait1,res1,stop1,start2,wait2,res2,stop2,post_start,post_stop = ana_df['timestamp']    
    ana_ov = extract_time_date(start1,wait1,res1,stop1) + extract_time_date(start2,wait2,res2,stop2) + extract_time_date(post_start,post_stop)
    
    #print sim_ov + ana_ov
    return sim_ov + ana_ov + preloop_ov
```

```python
def extract_timing_info(df,cores,inst):
    
    preloop_data,sim_data,ana_data = extract_data_time(df,cores,inst)
    preloop_exec,sim_exec,ana_exec = extract_exec_time(df,cores,inst)
    
    return [preloop_data,sim_data, ana_data],[preloop_exec,sim_exec, ana_exec]
```

```python
def extract_data_time(df,cores,inst):
    req_df = df[['step','StagingInput','Allocating','AgentStagingOutputPending','Done']]
    
    preloop_df = req_df[:1]
    sim_df = req_df[1:1+inst]
    ana_df = req_df[1+inst:]
    
    #Preloop extraction
    preloop_data_time  = list()
    for row in preloop_df.iterrows():
        step,t1,t2,t3,t4 = row[1:][0]
        preloop_data_time.append(extract_time(t1,t2,t3,t4))
        
    preloop_data_time = np.average(preloop_data_time)
        
    #Sim extraction
    sim_data_time  = list()
    for row in sim_df.iterrows():
        step,t1,t2,t3,t4 = row[1:][0]
        sim_data_time.append(extract_time(t1,t2,t3,t4))
        
    sim_data_time = np.average(sim_data_time)
    
    #Step2 extraction
    ana_data_time  = list()
    for row in ana_df.iterrows():
        step,t1,t2,t3,t4 = row[1:][0]
        #print t1,t2,t3,t4
        ana_data_time.append(extract_time(t1,t2,t3,t4))
    
    ana_data_time = np.average(ana_data_time)
    
    
    #print step1_data_time,step2_data_time
    return preloop_data_time,sim_data_time,ana_data_time
```

```python
def extract_exec_time(df,cores,inst):
    req_df = df[['step','Executing','AgentStagingOutputPending']]
    
    preloop_df = req_df[:1]
    sim_df = req_df[1:1+inst]
    ana_df = req_df[1+inst:]
    
    #Preloop extraction
    preloop_exec_time  = list()
    for row in preloop_df.iterrows():
        step,t1,t2 = row[1:][0]
        preloop_exec_time.append(extract_time(t1,t2))
        
    preloop_exec_time = np.average(preloop_exec_time)
    
    #Step1 extraction
    sim_exec_time  = list()
    for row in sim_df.iterrows():
        step,t1,t2 = row[1:][0]
        sim_exec_time.append(extract_time(t1,t2))
        
    sim_exec_time = np.average(sim_exec_time)
    
    #Step2 extraction
    ana_exec_time  = list()
    for row in ana_df.iterrows():
        step,t1, t2 = row[1:][0]
        ana_exec_time.append(extract_time(t1,t2))
    
    ana_exec_time = np.average(ana_exec_time)
    
    
    #print step1_data_time,step2_data_time
    return preloop_exec_time,sim_exec_time,ana_exec_time
```

```python
def separate(comp_df,err_df):
    ov_df = pd.DataFrame(columns=['EnMD Core overhead',
                                'EnMD Pat overhead',
                                'RP overhead'])
    ov_df_err = pd.DataFrame(columns=['EnMD Core overhead',
                                'EnMD Pat overhead',
                                'RP overhead'])
    ov_df['EnMD Core overhead'] = comp_df['EnMD Core overhead']
    ov_df['EnMD Pat overhead'] = comp_df['EnMD Pat overhead']
    ov_df['RP overhead'] = comp_df['RP overhead']
    
    ov_df_err['EnMD Core overhead'] = err_df['EnMD Core overhead']
    ov_df_err['EnMD Pat overhead'] = err_df['EnMD Pat overhead']
    ov_df_err['RP overhead'] = err_df['RP overhead']
    
    
    
    app_df = pd.DataFrame(columns=['preloop execution time',
                                   'sim execution time',
                                    'ana execution time'])
    app_df_err = pd.DataFrame(columns=['preloop execution time',
                                       'sim execution time',
                                        'ana execution time'])

    app_df['preloop execution time'] = comp_df['preloop execution time']
    app_df['sim execution time'] = comp_df['sim execution time']
    app_df['ana execution time'] = comp_df['ana execution time']
    
    app_df_err['preloop execution time'] = err_df['preloop execution time']
    app_df_err['sim execution time'] = err_df['sim execution time']
    app_df_err['ana execution time'] = err_df['ana execution time']
    
    
    
    data_df = pd.DataFrame(columns=['data movement'])
    data_df_err = pd.DataFrame(columns=['data movement'])
    data_df['data movement'] = comp_df['data movement']
    data_df_err['data movement'] = err_df['data movement']
    
    split_plot([ov_df,ov_df_err],[app_df,app_df_err],[data_df,data_df_err])
```

```python
def extract_rp_overhead(df,data,exec_list):
    
    
    preloop_df = df[:4]
    sim_df = df[4:10]
    ana_df = df[10:]
    
    start,wait,res,stop = preloop_df['timestamp']
    preloop_wait = extract_time_date(wait,res)
    preloop_rp_ov = preloop_wait - data[0] - exec_list[0]
    
    start1,wait1,res1,stop1,post_start,post_stop = sim_df['timestamp']    
    sim_wait = extract_time_date(wait1,res1)
    sim_rp_ov = sim_wait - data[1] - exec_list[1]
    
    start1,wait1,res1,stop1,start2,wait2,res2,stop2,post_start,post_stop = ana_df['timestamp']    
    ana_wait = extract_time_date(wait1,res1) + extract_time_date(wait2,res2)
    ana_rp_ov = ana_wait - data[2] - exec_list[2]
    
    rp_ov = preloop_rp_ov + sim_rp_ov + ana_rp_ov
    
    
    #print preloop_rp_ov, sim_rp_ov, ana_rp_ov
    
    return rp_ov
```

```python
def split_plot(ov,app,data):
    fig, axes = plt.subplots(nrows=3, ncols=1)
    ax1 = ov[0].plot(ax=axes[1],kind='bar',stacked=False,
               title='Overhead: Gromacs-LSDMap MD application implemented with Simulation-Analysis-Loop pattern on XSEDE.Comet',
               colors=['black','green','red'],ylim=(0,50),fontsize=18,rot=0,yerr=ov[1])
    ax2 = app[0].plot(ax=axes[0],kind='bar',stacked=False,
               title='Application: Gromacs-LSDMap MD application implemented with Simulation-Analysis-Loop pattern on XSEDE.Comet',
               colors=['cyan','blue','yellow'],ylim=(0,80),fontsize=18,rot=0,yerr=app[1])
    ax3 = data[0].plot(ax=axes[2],kind='bar',stacked=False,
               title='Data Movement: Gromacs-LSDMap MD application implemented with Simulation-Analysis-Loop pattern on XSEDE.Comet',
               colors=['brown'],ylim=(0,100),fontsize=18,rot=0,yerr=data[1])
    
    ax2.set_xlabel('Cores/No. of simulations',fontsize=18)
    ax2.set_ylabel('Time (seconds)',fontsize=18)
    ax2.legend(fontsize=18,loc='upper left')
    ax2.set_title(ax2.get_title(),fontsize=18)
    ax1.set_xlabel('Cores/No. of simulations',fontsize=18)
    ax1.set_ylabel('Time (seconds)',fontsize=18)
    ax1.legend(fontsize=18,loc='upper left')
    ax1.set_title(ax1.get_title(),fontsize=18)
    ax3.set_xlabel('Cores/No. of simulations',fontsize=18)
    ax3.set_ylabel('Time (seconds)',fontsize=18)
    ax3.legend(fontsize=18,loc='upper left')
    ax3.set_title(ax3.get_title(),fontsize=18)
```

```python
#MAIN

comp_df = pd.DataFrame(columns=['EnMD Core overhead',
                                'EnMD Pat overhead',
                                'RP overhead',
                                'preloop execution time',
                                'sim execution time',
                                'ana execution time',
                                'data movement'])

err_df = pd.DataFrame(columns=['EnMD Core overhead',
                                'EnMD Pat overhead',
                                'RP overhead',
                                'preloop execution time',
                                'sim execution time',
                                'ana execution time',
                                'data movement'])

for c in cores:
    
    core_ov_list = list()
    pat_ov_list = list()
    rp_ov_list = list()
    data_list = list()
    preloop_exec_list = list()
    sim_exec_list = list()
    ana_exec_list = list()
    
    for t in range(1,trials+1):
        #f_core = open('experiment_iter{0}_p{1}_i{1}/enmd_core_overhead.csv'.format(t,c))
        df_core = pd.read_csv('experiment_iter{0}_p{1}_i{1}/enmd_core_overhead.csv'.format(t,c),header=0,sep=',')
        df_pat = pd.read_csv('experiment_iter{0}_p{1}_i{1}/enmd_pat_overhead.csv'.format(t,c),header=0,sep=',')
        
        fname = glob.glob('experiment_iter{0}_p{1}_i{1}/execution*.csv'.format(t,c))[0]
        df_cu = pd.read_csv('{0}'.format(fname),header=0,sep=',',skipinitialspace=True)
        core_ov = extract_core_overhead(df_core)
        pat_ov = extract_pat_overhead(df_pat)
        data, exec_list = extract_timing_info(df_cu,cores=c,inst=c)
        rp_ov = extract_rp_overhead(df_pat,data,exec_list)
        
        
        core_ov_list.append(core_ov)
        pat_ov_list.append(pat_ov)
        rp_ov_list.append(rp_ov)
        preloop_exec_list.append(exec_list[0])
        sim_exec_list.append(exec_list[1])
        ana_exec_list.append(exec_list[2])
        data_list.append(sum(data))
 
    comp_df.loc['{0}/{1}'.format(c,c)] = [np.average(core_ov_list),
                                          np.average(pat_ov_list),
                                          np.average(rp_ov_list),
                                          np.average(preloop_exec_list),
                                          np.average(sim_exec_list),
                                          np.average(ana_exec_list),
                                          np.average(data_list)]
    
    err_df.loc['{0}/{1}'.format(c,c)] = [
                                          np.std(core_ov_list)/math.sqrt(trials),
                                          np.std(pat_ov_list)/math.sqrt(trials),
                                          np.std(rp_ov_list)/math.sqrt(trials),
                                          np.std(preloop_exec_list)/math.sqrt(trials),
                                          np.std(sim_exec_list)/math.sqrt(trials),
                                          np.std(ana_exec_list)/math.sqrt(trials),
                                          np.std(data_list)/math.sqrt(trials)
                                         ]
    
        
separate(comp_df,err_df)
```
